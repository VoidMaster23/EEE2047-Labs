## Basic Jupyter usage

A notebook is basically composed of cells.  A cell can contain text ("markdown") or code, in this case Python.  For any cell you can select the content type using the pulldown on the menubar.  A cell can be executed by pressing shift-enter while it is in focus.

Below is a cell that creates two numpy arrays.  The values in `xv` are initialised to be a particular function of `tv`.  Select it and press shift-enter to run the code in the Python kernel attached to the notebook.  Change the number of points in `tv` and press shift-enter again to run, and note the change in output.

```python
import numpy as np

tv = np.linspace(0, 10, 10);
xv = 0.1*tv**2 - np.cos(tv);
print(xv);
```

The Pythonn kernel is fully-featured for numeric computation, and includes most of the packages useful for scientific computing.  The following code plots the vectors defined above using "matplotlib".  The `%matplotlib notebook` line makes any plots appear in the notebook itself rather than in a separate window.

```python
import matplotlib.pyplot as plt
%matplotlib notebook

plt.plot(tv, xv, 'ro-')
plt.xlabel('t');  plt.ylabel('x(t)');
```

When you execute the code above it runs it in the Python kernel for the notebook, which already has the variables `tv` and `xv` defined. # Question 1

Create a plot of $x^2sin(1/x^2)+x$ on the interval [−1,1] using 250 points. Remember to label the axes.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook

x = np.linspace(-1,1,250)
y = ((x**2)*np.sin(1/(x**2)))+x

fig, ax = plt.subplots()
ax.plot(x, y)
plt.ylabel('f(x)')
plt.xlabel('x')
ax.set_title('Question 1')


```

# Question 2

We often need to plot complex-valued signals, where for each value of  𝑡  there is a real and an imaginary value, or a magnitude and a phase. We therefore need two sets of axes. Use the subplot functionality of matplotlib.pyplot to plot the real and imaginary parts of the signal  $ x(t) = e^{j \omega_0 t} $ in a single figure, but on two separate axes. Use a value of  $ \omega_0 = 2 $  and make the plot range over $ t=0 $ to $ t=10 $.

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0,10,250)

fx, (ax1, ax2) = plt.subplots(2, 1)

#Real numbers
ax1.set_title('t vs x(t) real')
ax1.set_xlabel('t')
ax1.set_ylabel('x(t) real')
ax1.plot(t, np.cos(2*t))

#Imaginary numbers
pos1 = ax2.get_position()
pos2 = [pos1.x0,pos1.y0 + 1, pos1.width, pos1.height]
ax2.set_position(pos2)
ax2.set_title('t vs x(t) imaginary')
ax2.set_xlabel('t')
ax2.set_ylabel('x(t) imaginary')
ax2.plot(t, np.imag(1j*np.sin(2*t)), 'g' )
```

# Question 3

 Most of signal processing involves applying a recursive relationship to some given data.  This course concentrates on continuous-time signals, but equivalent processes can be applied in the digital case.  Consider the recursive equation $ x[n] = -0.98 x[n-1] $.  Create a numpy array with 100 elements for each of the values $ x[0] $ to $ x[99] $, and write code to populate it assuming the initial condition $ x[0] = 10 $.  Use `stem` to make a plot of $ x[n] $ versus $ n $ over the range calculated. 

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.empty(100)
x = np.arange(100)


def recurse(n):
    if n == 0:
        return 10
        
    else:
        
        return -0.98*recurse(n-1)
    

for i in range(99,-1,-1):
    y[i] = recurse(i)




plt.title('Stem Plot of x[n] vs n')
plt.xlabel('n')
plt.ylabel('x[n]')

plt.stem(x,y,'g')  


```

# Question 4

 Create a surface plot of $ \sin(x) \sin(y) $ on the domain $ [-\pi,\pi] \times [-\pi,\pi] $

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D

def fun(x, y):
    return np.sin(x)*np.sin(y)



x= y= np.linspace(-np.pi, np.pi, 250)

X, Y = np.meshgrid(x, y)

fig = plt.figure()
ax = fig.gca(projection ='3d')
Z = fun(X,Y)
ax.set_ylabel('y')
ax.set_xlabel('x')
ax.set_zlabel('z')
ax.set_title('Surface plot of sin(x)sin(y)')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='viridis', 
                       edgecolor='none')



fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
